IceCream Seller Revenue Prediction based on Temperature

Dataset Link: https://github.com/Shahid-Fakhri/AI-Course

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Dataset
df = pd.read_csv('IceCreamData.csv')
df.head()
```

```python
# Check Summary
df.info()
```

```python
# Check Missing Values
df.isna().sum()
```

```python
# Visualization
sns.pairplot(df)
```

```python
# Features and Target
X = df[['Temperature']]
y = df['Revenue']
```

```python
# Train and test data split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
X_train.shape
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_model = LinearRegression()
```

```python
# Model Training
reg_model.fit(X_train, y_train)
```

```python
# Model Testing
y_predict = reg_model.predict(X)
```

```python
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
# Compare Actual and Prtedicted values

compare_df = pd.DataFrame({
    'Actual Revenue': y,
    'predicted Revenue':y_predict
})
compare_df
```

Model Evaluation

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y,y_predict)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X_test, y_test, color='blue')
plt.plot(X, y_predict, color='red')
plt.legend(['Original', 'Predicted Data'])
plt.legend()
plt.show()
```

```python
reg_model.score(X_train, y_train)
```

```python
reg_model.score(X_test,y_test)
```
